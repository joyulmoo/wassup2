# Q1-3
어느 지역에서 제일 많은 확진자가 발생했는지 1~5위 까지 출력하여라 (거주지 기준)

````python
df_group = df_covid.groupby("거주지").count()
df_group
````

````python
df_group = df_group.sort_values("확진일", ascending=False)
df_group
````

````python
top5= list(df_group[:5].index.values)
top5
````


# Q1-4
확진자가 가장 많이 발생한 1, 2, 3위 지역의 데이타만 출력하여라 

````python
for i in top5[:3]:
    print(df_covid[df_covid["거주지"]==i].sort_values("확진일"))
````
