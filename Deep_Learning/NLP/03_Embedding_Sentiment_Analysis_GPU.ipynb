# I. Naver Sentiment Movie Corpus v1.0

* train_data : 150,000
* test_data : 50,000

* DataFrame 합치기

# II. Data Preprocessing

> ## 1) 'document' 추출

```python
DF_X = DF.iloc[:, 1]
```

```python
DF_X[:4]
```

> ## 2) 문장 및 단어 크기 조정

* 문장을 단어별 나누기

```python
sentences = [sentence.split(' ') for sentence in DF_X]
```

* 처리 결과 확인

```python
sentences[:4]
```

* 문장을 단어 개수 순서로 정렬 후 시각화

```python
import matplotlib.pyplot as plt

sentence_len = [len(sentence) for sentence in sentences]
sentence_len.sort()

plt.plot(sentence_len)
plt.show()
```

* 32단어 이하의 문장 개수 확인

```python
print(sum([int(i <= 32) for i in sentence_len]))
```

* 문장 앞 32개 단어만 추출 후 단어 길이를 5글자로 자르기

```python
sentences_new = []

for sentence in sentences:
    sentences_new.append([word[:5] for word in sentence][:32])

sentences = sentences_new
```

> ## 3) 처리 결과 확인

```python
sentences[:4]
```

```python
sentence_len = [len(sentence) for sentence in sentences]
sentence_len.sort()

plt.plot(sentence_len)
plt.show()
```

> ## 4) 'label' - y

```python
y = DF.iloc[:, 2]
```

```python
len(y)
```

# III. 'tokenizer( )' and 'pad_sequences( )'

> ## 1) Tokenization & Vectorization

* '20000' 종류 단어 사용

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words = 20000)
tokenizer.fit_on_texts(sentences)
TV_X = tokenizer.texts_to_sequences(sentences)
```

```python
TV_X[:5]
```

> ## 2) pad_sequences( )

* 동일 길이(32) 벡터로 변경

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

TV_X = pad_sequences(TV_X,
                     maxlen = 32,
                     padding = 'post')
```

> ## 3) 처리 결과 확인

```python
print(TV_X[:5])
```

```python
TV_X.shape
```

> ## 4) train_test_split( )

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(TV_X,
                                                    y,
                                                    test_size = 0.3,
                                                    random_state = 2045)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

# IV. Modeling

> ## 1) Model Define

 - Embedding Dimension : 256
    - Input_dim: 적용 단어 종류
    - output_dim: 임베딩 차원

```python
from tensorflow.keras import models, layers

model = models.Sequential()

model.add(layers.Embedding(input_dim = 20000, # input size는 32 이고 2만개의 단어 종류로 이루어짐
                           output_dim = 256)) # word2vec size

model.add(layers.LSTM(32))
model.add(layers.Dense(1, activation = 'sigmoid')) #긍부정 분류를 위한 레이어

model.summary()
```

> ## 2) Model Compile

```python
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
```

> ## 3) Model Fit

* 약 4분

```python
%%time

history = model.fit(X_train, y_train,
                    epochs = 20,
                    batch_size = 128,
                    validation_data = (X_test, y_test))
```

> ## 4) Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label = 'loss')
plt.plot(history.history['val_loss'], 'r--', label = 'val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label = 'accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label = 'val_accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```

> ## 5) Model Evaluate

```python
model.evaluate(X_test, y_test, verbose = 0)
```

# V. 문장 감성 분석

* 입력 단어에 따라서 감성 분석 결과 변화
 - 긍정('1') -> 부정('0')

```python
test_sentence = '재미있을 줄 알았는데 완전 실망했다. 너무 졸리고 돈이 아까웠다.'
test_sentence = test_sentence.split(' ')

test_sentences = []
now_sentence = []

for word in test_sentence:
    now_sentence.append(word)
    test_sentences.append(now_sentence[:])

test_X_1 = tokenizer.texts_to_sequences(test_sentences)
test_X_1 = pad_sequences(test_X_1, padding = 'post', maxlen = 32)

prediction = model.predict(test_X_1, verbose = 0)

for idx, sentence in enumerate(test_sentences):
    print(sentence)
    print(prediction[idx])
```
